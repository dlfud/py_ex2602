```python
# 클로져 
# - 안쪽 함수가 바깥 변수 값을 기억하는 구조
# - 주로 '상태를 유지하는 함수'를 만들 때 사용됩니다.

# 함수안에 함수 만드는것이 가능. return으로 꺼낼 수 있음
def make_plus(x):
    def plus(n):
        return x + n
    return plus 
rs = make_plus(3) # x가 3이라는 것을 기억함. plus라는 함수 생성됨. rs변수에는 리턴값인 plus함수가 담김
print(rs(5)) # n이 5가 됨. 
# return 에서 plus를 반환하므로 rs에는 plus가 저장됨. 사실상 print(plus(5))
```

```python
# 데코레이터
# - 기존 함수를 고치지 않고, 기능만 덧붙이는 방법
# - 내부 구조는 클로저
# - 로그 출력, 실행 시간 측정 등에 사용됩니다.

def deco(fun):
    def wrap():
        print("실행 전")
        fun()
        print("실행 후")
    return wrap

# def hello():
#     print("안녕하세요")    
# new_fun = deco(hello)
# new_fun()

#어노테이션 사용
# @deco / def hello(): -> hello = deco(hello)
@deco
def hello():
    print("안녕하세요")
hello() # hello함수는 wrap으로 바뀌게 됨
```

```python
# 어노테이션
# - 어떤 타입인지 미리 적어두는 표시
# - 어노테이션은 강제 규칙이 아니다
# - 어노테이션은 '설명용'입니다.
# - 파이썬이 강제로 막아주지는 않습니다.
# - 결과가 int일거다 라는 빠른 판단이 가능해서 사용
age: int = 30
print(age)
age: int = "안녕"
print(age)

# a는 int, b는 int, 결과도 int
# int라고 썼는데 str써도 에러 안남
def add(a:int, b:int) -> int:
    return a + b
# add(10, 20) #30
add("안녕", "반가워")

numbers: list[int] = [1, 2, 3] # 리스트
person: dict[str, int] = { # 딕셔너리
    "age": 30,
    "height": 170
}

print(numbers)
print(person)
```

```python
# 정규표현식(정규식)

text = "안녕하세요. 저는 15살 입니다. 저는 안경이 12개 있습니다."
print("15" in text) # 15라는 글자를 찾고 있으면 True 반환. 간단하게 찾으려면 이렇게 사용

import re
# \d : 숫자
# + : 1개 이상
pattern = r"\d+" # r은 raw string. \(역슬래시) 특수문자
result = re.findall(pattern, text)
print(result)

sample = "abc123xyz456"
re.findall(r"\d+", sample) # ['123', '456']
re.findall(r"[a-z]+", sample) # ['abc', 'def']


# 특정 형식 찾기
data = "test123@gmail.com"
# \w : 문자 또는 숫자
# + : 한개 이상
# @ : 문자 그대로 @
# \. : 점(특수 문자이기 때문에 \ 붙임)
email_pattern = r"\w+@\w+\.\w+"
emails = re.findall(email_pattern, data)
print(emails)

# 문자열 치환
test = "가격은 1200원 입니다."
# re.sub(패턴, 바꿀값, 원본문자열)
new_text = re.sub(r"\d+", "", test)
print(new_text)

# 정규식 vs 일반 문자열 처리
test = "뭄무게: 80kg, 키: 180"
# - split은 구분자가 명확할 때 유리합니다.
# - 정규식은 '형식이 일정한 패턴'을 찾을 때 유리합니다.
test.split(",")
numbers = re.findall(r"\d+", test)
print(numbers)
```
