```python
# 클로져 
# - 안쪽 함수가 바깥 변수 값을 기억하는 구조

# 함수안에 함수 만드는것이 가능. return으로 꺼낼 수 있음
def make_plus(x):
    def plus(n):
        return x + n
    return plus 
rs = make_plus(3) # x가 3이라는 것을 기억함. plus라는 함수 생성됨. rs변수에는 리턴값인 plus함수가 담김
print(rs(5)) # n이 5가 됨. 
# return 에서 plus를 반환하므로 rs에는 plus가 저장됨. 사실상 print(plus(5))
```

```python
# 데코레이터
# - 기존 함수를 고치지 않고, 기능만 덧붙이는 방법
# - 내부적으로 클로저를 사용함

def deco(fun):
    def wrap():
        print("실행 전")
        fun()
        print("실행 후")
    return wrap

# def hello():
#     print("안녕하세요")    
# new_fun = deco(hello)
# new_fun()

#어노테이션 사용
# hello = deco(hello) -> @deco /  def hello():부분
@deco
def hello():
    print("안녕하세요")
hello() # hello함수는 wrap으로 바뀌게 됨
```
