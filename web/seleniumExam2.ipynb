```python
# 셀레니움
# 요소 찾는 방법 3가지
# - 태그이름(TAG_NAME)
# - class / id (CLASS_NAME / ID)
# - xPath

from selenium import webdriver
from selenium.webdriver.common.by import By

# ========================================================
# 0) 브라우저 실행 + 페이지접속
driver = webdriver.Edge()
driver.get('https://quotes.toscrape.com/')

# 수집한 결과를 담을 리스트
results = []

# ========================================================
# find_element vs find_elements
# find_element : 요소 1개만 찾음 -> 없으면 에러 발생
# find_elements : 0개 이상을 반환 -> 없으면 빈 리스트 반환
one_quote_block = driver.find_element(By.CLASS_NAME, 'quote')
many_quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')
print("find_element로 찾은 quote 블록 1개 :", type(one_quote_block))
print("find_elements로 찾은 quote 블록 개수 :", len(many_quote_blocks))

# ========================================================
# 1) 태그이름(TAG_NAME)으로 요소 찾기
# - 범위가 너무 넓어지는 경우가 많아서 실무에서는 연습용으로 사용
div_elements = driver.find_elements(By.TAG_NAME, "div")
print("TAG_NAME으로 찾은 div 태그 개수 :", len(div_elements)) # 디버깅이나 검증용으로도 사용

# ========================================================
# 2) class로 요소 찾기(CLASS_NAME)
# - selenium에서 가장 자주 사용하는 방식 중 하나
# - class에 공백이 포함된 경우(클래스가 2개 이상 있을 경우 class="a b") : CLASS_NAME으로 바로 찾기 어려움
first_quote_text = driver.find_element(By.CLASS_NAME, 'text')
print("CLASS_NAME으로 찾은 첫 번째 명언 :", first_quote_text.text)

# ========================================================
# 3) id로 요소 찾기(ID)
# - id는 페이지에서 '유일'하게 존재 
# - 정확하게 찾고 싶을 때 가장 안정적
quotes_container = driver.find_elements(By.ID, 'quotes')
# print("ID로 찾은 요소의 태그 이름 :", quotes_container[0].tag_name) # 빈 리스트이므로 [0]해도에러
if quotes_container:
    print("태그 이름:", quotes_container[0].tag_name)
else:
    print("해당 ID를 가진 요소가 없습니다.")

# ========================================================
# 4) xPath로 요소 찾기
# - HTML구조(경로)를 기준으로 요소를 찾는 방식
# - class, id가 애매하거나 조건이 복잡할 때 유용
# - 장점 : 조건/경로 기반으로 정확하게 찾기 쉬움
# - 단점 : HTML구조가 바뀌면 xPath도 함께 깨질 수 있음

# // : 문서 전체에서 검색(root부터 전역 탐색)
# div[@class="quote"] : div태그 중 class="quote"인 요소
# span[@class="text"] : 위에서 찾은 div 바로 아래에 있는 span태그 중 class="text"인 요소
xpath_quote_text = driver.find_element(
    By.XPATH, 
    "//div[@class='quote']/span[@class='text']"
)
print("XPath로 찾은 명언 :", xpath_quote_text.text)

# 첫번째 quote 블록만 특정해서 찾는 XPath예시
xpath_first_block = driver.find_element(By.XPATH, "(//div[@class='quote'])[1]")
xpath_first_author = xpath_first_block.find_element(By.CLASS_NAME, "author")
print("XPath로 찾은 첫번쨰 저자 :", xpath_first_author.text)


# 5) 브라우저 종료
driver.quit()
```
