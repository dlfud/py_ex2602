```python
# requests 모듈 : 웹 페이지를 요청하고 응답 데이터를 받을 수 있음.
import requests as req
from bs4 import BeautifulSoup # BeautifulSoup -> html해석기

# req 요청시 오류 났을 때 헤더정보를 크롬으로 지정
# request_headers = {'User-Agent': ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98Safari/537.36'), }

response = req.get("https://www.naver.com/")
# print(response) # 200이 떠야함

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

# print(html_doc)

# soup을 이용해서 html_doc 문자열 파싱, 그 내용을 soup 변수에 대입
soup = BeautifulSoup(html_doc, 'html.parser') # 문자열 쪼개서 soup변수에 넣음

a_list = soup.select('a')
# a태그들 리스트
print("===a태그 리스트===")
print(a_list)
print("===a태그 리스트 중 첫번째===")
print(a_list[0])
# a태그 안에 내용만 가져오기
print("===a태그 리스트 중 첫번째 텍스트===")
print(a_list[0].text) # Elsie
print("===a태그 리스트 중 첫번째 링크===")
print(a_list[0]['href']) # 속성을 선택할 땐 대괄호
# for문 돌려서 내용만 가져오기
print("===a태그 리스트에 있는 텍스트 모두 출력===")
for text in a_list:
    print(text.text)
# for문 돌려서 링크출력
print("===a태그 리스트에 있는 링크 모두 출력===")
for link in a_list:
    print(link['href']) # 속성을 선택할 땐 대괄호
```
