```python
# selenium: 브라우저를 사람이 조작하듯이 자동으로 조작할 수 있는 도구

# 브라우저를 실행하고 조작하는 객체 
from selenium import webdriver
# 요소를 찾을 때 사용하는 선택 방식 지정 도구
from selenium.webdriver.common.by import By

# 1. 브라우저 실생
# driver = webdriver.Chrome() # 실행 시 크롬 열림
driver = webdriver.Edge() # 엣지 실행

# 2. 웹페이지 접속
# .get('url') : 특정 웹 페이지로 이동하는 명령어
driver.get('https://quotes.toscrape.com/') # 엣지로 해당 url열림

# 3. 요소 하나 찾기
# find_element : 웹페이지에서 엘리먼트 찾음
# 'text'이름의 클래스 요소를 찾겠다
quote = driver.find_element(By.CLASS_NAME, 'text')

# 4. 텍스트 가져오기
print(quote.text)

# 5. 브라우저 종료
driver.quit()
```
