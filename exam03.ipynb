```python
def is_prime_number(num):
    if num <= 1:
        return False
    
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print("1은 소수입니다 : {}".format(is_prime_number(1)))
print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6는 소수입니다 : {}".format(is_prime_number(6)))
print("7는 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```

```python
def is_prime_number(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def get_1_to_n_prime_sum(n):
    sum = 0
    for i in range(1, n + 1):
        if is_prime_number(i):
            sum += i
    return sum

n = 10
print("1부터 {}사이에 존재하는 모든 소수 의 합 : {}".format(n, get_1_to_n_prime_sum(n)))
```

```python
month_end_day = {
    '1월': 31,
    '2월': 28,
    '3월': 31
}
print(month_end_day)

for i in range(1, 12 + 1):
    month = str(i) + '월'
    end_day = 31
    if i == 2:
        end_day = 28
    elif i in [4, 6, 9, 11]:
        end_day = 30
    month_end_day[month] = end_day
print(month_end_day)
```
