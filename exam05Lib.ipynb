```python
# 데이터 형식 모듈
# JSON 형식으로 데이터 변환
# - JSON : 데이터를 약속된 모양으로 적는 방식("key": "value")
# - 딕셔너리와 모양이 같음. 딕셔너리와 다를게 없음
# 웹, API, 파일 저장에 사용
# - API : 프로그램끼리 대화하는 방법(규칙)
import json

data = {
    "name": "jin",
    "age": 20
}

print(json.dumps(data))
```

```python
# 정규 표현식 모듈 (정규식 모듈)
# 문자열 패턴 검사 모듈
# 형식 확인(아이디, 이메일 등)에 사용
import re
# 문자열이 hi로 시작하는지 확인
print(bool(re.match("hi", "hi python")))
```

```python
# CSV 파일 모듈
# CSV(엑셀) 파일을 읽고 쓰는 모듈
# 데이터 저장/분석 기초에 사용
import csv
# csv 파일에 한줄 데이터 저장
# 파일 열고 자동으로 닫는 구문
# 파일 이름, w : write, 줄바꿈 방지
# 쓰기 모드로 파일 열고, f라는 이름으로 사용, 종료 시 자동 닫기, 줄바꿈 문제 방지
with open("test.csv", 'w', newline='') as f:
    csv.writer(f).writerow(['name', 'score']) #writerow : name, score 가로로 적겠다
```
