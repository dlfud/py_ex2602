```python
answer = 0
n = 13
w = 3
num = 6
h = int((n / w) + 1)
boxs = []
x = 1

print(h)
for i in range(0, h):
    t = []
    for j in range(w):
        if x <= n:
            t.append(x)
            x += 1
        else:
            t.append(x)
        
    if i % 2 == 0 :
        boxs.append(t)
    else:
        t.reverse()
        boxs.append(t)

print(boxs)

for i in range(len(boxs)):
    for j in range(len(boxs[i])):
        if boxs[i][j] == num:
            d = i
            while d < h and boxs[d][j]: # 높이보다 현재 위치가 작고, 그자리에 박스가 있다면 (0이 아니라면)
                answer += 1
                d += 1
    
    
print(answer)
```
