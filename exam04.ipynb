```python
s = "안녕하세요."
print(len(s))
print("s를 이루는 문자 앞에거 2개 출력")
print(s[0:2])
print("s를 이루는 문자 하나 건너뛰고 1개 출력")
print(s[1:2])
print("s를 이루는 문자 하나 앞에서 4개 출력")
print(s[:4])
print("s를 이루는 문자 하나 3개 건너뛰고 나머지 출력")
print(s[3:len(s)])
print(s[3:])
```

```python
s = "안녕하세요"
print("s를 이루는 문자를 하나씩 출력, while")
i = 0
while i < len(s):
    print(s[i])
    i += 1

print("s를 이루는 문자를 역순으로 하나씩 출력, while")
i = len(s) - 1
while i >= 0:
    print(s[i])
    i -= 1

print("s를 이루는 문자를 역순으로 하나씩 출력, for")
for c in s:
    print(c)
```

```python
s = "a안b녕c하d세e요f"
print("s를 이루는 문장 중 한글만 출력")
print(s[1::2])
print("for, if, or, continue사용")
for c in s:
    if c == 'a' or c == 'b' or c == 'c' or c == 'd' or c == 'e' or c == 'f':
        continue
    print(c)
print("for, if, and, continue사용")
for c in s:
    if c != 'a' and c != 'b' and c != 'c' and c != 'd' and c != 'e' and c != 'f':
        print(c)

print("for, not in사용")
for c in s:
    if c not in ['a', 'b', 'c', 'd', 'e', 'f']:
        print(c)
```
