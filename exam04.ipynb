```python
s = "안녕하세요."
print(len(s))
print("s를 이루는 문자 앞에거 2개 출력")
print(s[0:2])
print("s를 이루는 문자 하나 건너뛰고 1개 출력")
print(s[1:2])
print("s를 이루는 문자 하나 앞에서 4개 출력")
print(s[:4])
print("s를 이루는 문자 하나 3개 건너뛰고 나머지 출력")
print(s[3:len(s)])
print(s[3:])
```

```python
s = "안녕하세요"
print("s를 이루는 문자를 하나씩 출력, while")
i = 0
while i < len(s):
    print(s[i])
    i += 1

print("s를 이루는 문자를 역순으로 하나씩 출력, while")
i = len(s) - 1
while i >= 0:
    print(s[i])
    i -= 1

print("s를 이루는 문자를 역순으로 하나씩 출력, for")
for c in s:
    print(c)
```

```python
s = "a안b녕c하d세e요f"
print("s를 이루는 문장 중 한글만 출력")
print(s[1::2])
print("for, if, or, continue사용")
for c in s:
    if c == 'a' or c == 'b' or c == 'c' or c == 'd' or c == 'e' or c == 'f':
        continue
    print(c)
print("for, if, and, continue사용")
for c in s:
    if c != 'a' and c != 'b' and c != 'c' and c != 'd' and c != 'e' and c != 'f':
        print(c)

print("for, not in사용")
for c in s:
    if c not in ['a', 'b', 'c', 'd', 'e', 'f']:
        print(c)
```

```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종
# 프로그래밍으로 객체를 만들 때 사용하는것

# 전사, 법사, 궁수, 자동차, 비행기 -> object
# -> 현실, 가상, 개념

# 속성 (저장, 명사적) - attribute (모리모양, 옷, ...)
# 능력 (기능, 동사적) - method (공격하다, 방어하다, ...)

class Person:
    def hello(self): # 메서드는 항상 매개변수 자리에 self가 있어야 함. 규칙임
        print("hi")

jin = Person() # jin은 Person의 인스턴스
jin.hello()
```

```python
class Person:
    def hello(self):
        print("hello")
    def hi(self):
        self.hello() # self를 통해 메소드 접근 가능
jin = Person()
jin.hi()
```
