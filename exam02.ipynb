```python
a = 10

# `==` => 같다.
if ( a == 10 ):
  print("참")

# `!=` => 같지 않다.
if ( a != 10 ):
  print("거짓")

if ( a > 10 ):
  print("거짓")

if ( a >= 10 ):
  print("참")

b = 10

if ( a == b ):
  print("참")

# c 에는 true or false 이 담긴다.
c = a != b # False

if ( c ):
  print("f")

if ( c == False ):
  print("t")

# `not` => 반전
if ( not c ):
  print("t")

if ( not(not c) ):
  print("f")

d = True

if ( c != d ):
  print("t")

if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("f")

if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("t")
# 문제끝
```

```python
# 문제 : 10대, 20대, 30대 중 나이에 맞는 변수를 출력해주세요.

age = 32

print("== 정답 v1 ==")
if age >= 10 and age <= 19:
  print("10대")
if age >= 20 and age <= 29:
  print("20대")
if age >= 30 and age <= 39:
  print("30대")

print("== 정답 v2 ==")
if age >= 10 and age <= 19:
  print("10대")
elif age >= 20 and age <= 29:
  print("20대")
elif age >= 30 and age <= 39:
  print("30대")

print("== 정답 v3 ==")
if age >= 10 and age <= 19:
  print("10대")
elif age >= 20 and age <= 29:
  print("20대")
else:
  print("30대")

print("== 정답 v4 ==")
if age >= 10:
  if age <= 19:
    print("10대")

if age >= 20:
  if age <= 29:
    print("20대")

if age >= 30:
  if age <= 39:
    print("30대")

print("== 정답 v5 ==")
if age <= 19:
  print("10대")
elif age <= 29:
  print("20대")
elif age <= 39:
  print("30대")

print("== 정답 v6(깔끔) ==")
if age <= 19:
  print("10대")
elif age <= 29:
  print("20대")
else:
  print("30대")

print("== 정답 v7 ==")
if age >= 30:
  print("30대")
elif age >= 20:
  print("20대")
elif age >= 10:
  print("10대")

print("== 정답 v7(깔끔) ==")
if age >= 30:
  print("30대")
elif age >= 20:
  print("20대")
else:
  print("10대")
```
