```python
# pandas : 파이썬에서 데이터 분석과 데이터 처리를 위한 대표적인 라이브러리
# 주로 표 형태의 데이터 (엑셀, CSV, 데이터베이스 등)을 다루는데 사용
# 데이터 과학, 머신러닝, 금융분석, 통계 처리 등에서 매우 많이 사용

import pandas as pd

# DataFrame : 여러 개의 열(Column)으로 구성된 표 형태의 데이터
# "열 이름" : [값들]
# - 각 열에는 열 이름이 있음
# - 각 행에는 자동으로 행 번호(index)가 붙음

# 딕셔너리 형태로 DataFrame을 만듦
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
# 전체 출력
# print(df)

# 열 하나씩 출력
# 결과는 DataFrame이 아니라 Series 라는 1차원 자료구조로 나옴
# Series: 1차원 데이터 (열 하나)
# DataFrame : 여러 개의 Series가 모여서 만들어진 2차우너 표 형태의 데이터
print(df["Name"])
print(df["Age"])
print(df["Sex"])
```

```python
import pandas as pd
# 이름이 Age인 Series 생성
ages = pd.Series([22, 35, 58], name="Age")
print(ages)

# DataFrame 에서 특정 열 꺼내기
print(df["Age"])

# 기본 통계 함수
# max : 최댓값
print("최대 나이", df["Age"].max())
# min : 최소값
print("최소 나이", df["Age"].min())
# mean : 평균값
print("평균 나이", df["Age"].mean())

# 요약 통계 정보 보기
# describe : 숫자형 열에 대해 요약 통계 정보를 한번에 보여줌
print(df.describe())
#              Age
# count   3.000000  데이터 개수
# mean   38.333333  평균
# std    18.230012  표준 편차(데이터가 평균에서 얼마나 퍼져 있는지)
# min    22.000000  최솟값
# 25%    28.500000  1사분위수(Q1)
# 50%    35.000000  중앙값(median)
# 75%    46.500000  3사분위수(Q3)
# max    58.000000  최댓값

# DataFrame의 전체 요약 정보 
# info : DataFrame의 구조 정보를 보여줌
print(df.info())
```

```python
# DataFrame에서 특정 행을 선택하는 방법 2가지
# - iloc : "위치(position)"로 선택(0번째, 1번째 처럼 순서/번호로 접근)
# - loc : "라벨(label)"로 선택(index에 붙어있는 이름으로 접근)

# iloc : 인덱스의 '값'이 무든 상관없이, '몇번째 줄이냐'로 가져옴
# df.iloc[0] => 첫번째 행(0번 위치)
# 결과는 '한 행'이므로 Series(1차원) 형태로 나옴
print(df.iloc[0]) # 0번 위치 (첫 번째 행)
print(df.iloc[1]) # 1번 위치 (두 번째 행)
print(df.iloc[2]) # 2번 위치 (세 번째 행)

# loc : 인덱스 'index라벨'로 접근
# 현재는 iloc과 loc은 결과가 같음
# index를 '이름'으로 바꾸면 (Name열을 index로 지정) loc은 이름을 기준으로 접근 가능
# df.index = ['사람1', '사람2', '사람3']
# df.loc('사람1')로 접근
print(df.loc[0])
```

```python
# head() : 위쪽(처음 부분) 몇개의 행만 보여주는 함수
# tail() : 아리쪽(마지막 부분) 몇개의 행만 보여주는 함수
# 데이터가 클 때 일부분만 빠르게 확인하는 경우 자주 사용

# 상위 5개
print(df.head())
# head()에 숫자를 안쓰면 기본값은 5
# 위에서부터 5개의 행을 보여줌
# 데이터가 5개 보다 적으면 있는 만큼만 출력

# 상위 2개
print(df.head(2))
# head(2)는 위에서부터 2개의 행만 출력

# 하위 5개
print(df.tail())
# 숫자를 안쓰면 기본값은 5개
# 아래에서부터 5개의 행을 보여줌

# 하위 2개
print(df.tail(2))
# 마지막에서부터 2개의 행만 출력
# 최근 데이터나 마지만 데이터 확인할 때 자주 사용
```

```python
# 조건 필터링 (Boolean Indexing)

# 나이가 30이상인 사람만 보기
print(df[df["Age"] >= 30])
# 성별이 여성(female)인 행만 보기
print(df[df["Sex"]=="female"])
```

```python
# 새로운 열 추가
df["Survived"] = [1, 0, 1]
print(df)

# 열 이름 바꾸기
df = df.rename(columns={"Sex" : "Gender"})
print(df)

# 열 삭제하기 (axis=0 열 / axis=1 행)
df = df.drop("Survived", axis=1)
print(df)
```

```python
# 결측치(NaN) 처리
# 결측치 : 값이 비어있는 상태
# pandas에선 보통 NaN(Not of Number)로 표시
# NOone을 넣어도 pandas에선 자동으로 NaN으로 변환

print(df)
print(df.isnull())

df2 = df.copy()
# copy() : 원본 df를 복제 (복사본 만들기)
df2.loc[1, "Age"] = None
print(df2.isnull())
# df2의 1번 행(두번째 사람)의 Age열 값을 None으로 변경
# loc[행 라벨, 열 이름] 형태로 특정 위치의 값을 바꿀 수 있음

print(df2.dropna(axis=0)) # axis=0 행부분 제거
# axis=0 → 행을 삭제
# axis=1 → 열을 삭제
# dropna() ; 결측치가 있는 부분 제거

print(df2.fillna(0))
# fillna(i) : 결측치를 i값으로 채우기
```

```python
# 특정 열만 따로 채우기
# 결측치를 평균 값으로 채우기
print(df2["Age"].fillna(df["Age"].mean()))
# Age열만 선택해서 fillna적용
# NaN을 평균값으로 대체
# 데이터 분석에서 자주 쓰는 방식

# 결측치를 이전 값으로 채우기
print(df2.fillna(method='ffill'))
# method='ffill' : foward fill => 바로 위 (이전행) 의 값으로 NaN을 채움
# 시간순 데이터(날짜 데이터) 에서 많이 사용

# 결측치를 다음 값으로 채우기
print(df2.fillna(method='bfill'))
# method='bfill' : backward fill => 바로 아래 (다음행) 의 값으로 NaN을 채움
# fill과 반대 방향
```

```python
# axis	방향	의미
# axis=0	위 → 아래	행 방향으로 계산
# axis=1	왼쪽 → 오른쪽	열 방향으로 계산

# numpy에선 axis=0 은 행 -> 위에서 아래로 계산을 하니깐
# 1 2 3
# 4 5 6

# 1+4
# 2+5
# 3+6

# pandas에서 dropna는 삭제할 축 이므로 axis=0은 행(가로줄), axis=1은 열(column) (세로줄)
```
