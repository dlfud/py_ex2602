```python
class Person():
    def __init__(self, name, age, hobby):
        # init 안에 있는것들은 인스턴스 속성
        self.hello = "안녕하세요"
        self.name = name 
        self.age = age
        self.hobby = hobby
    
```

```python
#클래스 속성
class Person():
    folder = [] # 클래스 속성 -> 클래스에 속한다 (공공제) 모든 인스턴스에서 접근이 가능 

    def add_files(self, file):
        # self.folder.append(file)
        Person.folder.append(file) # 클래스 이름으로 클래스 속성에 접근 가능
    # 설계도에 file을 넣을수 있는 리스트 만듬
    # add_files통해서 받아온 파일을 하나씩 넣음

jin = Person()
jin.add_files('txt')

poll = Person()
poll.add_files('img')

print(jin.folder) # ['txt', 'img']
print(poll.folder) # ['txt', 'img']
# 인스턴스 속성이라면 각각 가지고 있는게 맞음
# 클래스 속성은 설계도에 있음 -> 설계도는 다같이 쓸수 있는 곳 -> 너도 나도 접근 가능한 공용 폴더라고 생각

# 인스턴스 속성은 각각 한명씩 각각의 TV를 보는것으로 생각
# 클래스 속성은 TV한개를 다 같이 보는 것
```

```python
# 인스턴스 속성
class Person():
    def __init__(self):
        self.folder = []
    def add_files(self, file):
        self.folder.append(file)
    
jin = Person()
jin.add_files('txt')

poll = Person()
poll.add_files('img')

print(jin.folder) # ['txt']
print(poll.folder) # ['img']
```

```python
# 비공개 클래스 속성
class Person():
    __weight = 78
    def print_weight(self):
        print(Person.__weight)

jin = Person()
jin.print_weight() #우회 가능

# print(jin.__weight) # 에러남 비공개 클래스 속성에 접근함으로
```
